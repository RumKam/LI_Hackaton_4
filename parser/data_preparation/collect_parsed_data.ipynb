```python
import pandas as pd


#–∑–∞–≥—Ä—É–∑–∏–º –¥–∞—Ç–∞—Å–µ—Ç—ã
set1 = pd.read_csv('set1.csv')
set2 = pd.read_csv('set2.csv')
set3 = pd.read_csv('set3.csv')
set4 = pd.read_csv('set4.csv')
set5 = pd.read_csv('set5.csv')
set6 = pd.read_csv('best_alla.csv')
```

```python
# –æ–ø—Ä–µ–¥–µ–ª–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–∏ 
def transform_data(df):
    '''–§—É–Ω–∫—Ü–∏—è –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–∞—Ç–∞—Å–µ—Ç.
    –î–∞–ª–µ–µ –∫–∞–∂–¥–∞–π –ø–æ—Å—Ç –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –≤ 
    –æ—Ç–¥–µ–ª—å–Ω—É—é —Å—Ç—Ä–æ–∫—É –Ω–æ–≤–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞. 
    –û—Å—Ç–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥—É–±–ª–∏—Ä—É—é—Ç—Å—è –ø–æ –≤–ª–∞–¥–µ–ª—å—Ü–∞–º 
    –ø–æ—Å—Ç–æ–≤. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π
    –¥–∞—Ç–∞—Å–µ—Ç'''
    
    url = []
    name = []
    job = []
    text = []
    reactions = []
    
    for i in range(len(df)):
        y = df['text'][i].split("',")
        x = df['reactions'][i].split("',")
        
        for j in range(len(y)):
            url.append(df['url'][i])
            name.append(df['name'][i])
            job.append(df['job'][i])
            reactions.append(df['reactions'][i])
            text.append(y[j])
            
    df_t = pd.DataFrame(
        {
            'url_user': url,
            'text_post': text,
            'job': job,
            'count_comments': '-1',
            'count_reactions': '-1',
            'react_per_user': reactions,
            'name': name
        }
    )
    
    return df_t
```

```python
#—Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –¥–∞—Ç–∞—Å–µ—Ç—ã
posts_set1 = transform_data(set1)
posts_set2 = transform_data(set2)
posts_set3 = transform_data(set3)
posts_set4 = transform_data(set4)
posts_set5 = transform_data(set5)
posts_set6 = transform_data(set6)
```

```python
#–ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
posts_set1.sample(5)
```

```python
#—Å–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ —Å –¥–∞—Ç–∞—Å–µ—Ç–∞–º–∏
frames = [posts_set1, posts_set2, posts_set3, posts_set4, posts_set5, posts_set6]

#–æ–±—ä–µ–¥–∏–Ω–∏–º –¥–∞—Ç–∞—Å–µ—Ç—ã
result = pd.concat(frames)

#—É–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏, –≥–¥–µ –Ω–µ—Ç –ø–æ—Å—Ç–æ–≤
result = result[result['text_post'] != '[]']

#—Å–±—Ä–æ—Å–∏–º –∏–Ω–¥–µ–∫—Å—ã
result = result.reset_index()

#—É–¥–∞–ª–∏–º –ª–∏—à–Ω–∏–π —Å—Ç–æ–ª–±–µ—Ü
result = result.drop('index', axis=1)
```

```python
#–ø—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
result
```
